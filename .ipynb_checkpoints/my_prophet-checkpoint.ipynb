```python
from pandas import read_csv
from pandas import to_datetime
from pandas import DataFrame
from pandas import merge
from fbprophet import Prophet
import matplotlib.pyplot as plt
import numpy as np
import datetime
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
```

```python
# load data
df = read_csv('../gen_data/daily_sales.csv', usecols=[1,2] ,header=0)
```

```python
# prepare expected column names
df.columns = ['ds', 'y']
df['ds'] = to_datetime(df['ds'])
df_log = df[['ds', 'y']]
df_log['y'] = np.log(df['y']+1)
#归一化
#df['y'] = (df['y'] - df['y'].mean()) / (df['y'].std())
train = df_log.drop(df_log.index[-91:])
```

```python
holidays = DataFrame({
  'holiday': 'playoff',
  'ds': to_datetime(['2020-01-06','2020-01-10','2020-02-28','2020-01-09','2020-03-08',#
                     '2020-06-01','2020-06-16','2020-05-05','2020-05-10','2020-06-15',#
                     '2020-06-17','2020-06-18','2020-07-08','2020-06-20','2020-06-19',#
                     '2020-09-09','2020-11-01','2020-11-02','2020-11-11','2021-01-20',#
                     '2021-01-23','2021-01-24','2021-01-25','2021-02-02','2021-05-10',#
                     '2021-06-01','2021-06-02','2021-06-03','2021-06-16','2021-06-06',#]),
                     #'2020-01-07','2020-04-28','2020-05-07',
                     #'2020-05-15','2020-06-03',#'2020-06-12','2020-07-19',
                     #'2020-09-11','2020-11-04','2020-12-11',
                     '2021-05-01','2021-05-10',#'2020-12-12','2021-01-21','2021-01-22',
                     '2021-06-18',#'2021-06-04',#]),#'2020-01-08','2020-03-05','2020-06-02','2021-06-17','2020-11-03','2020-12-10','2021-03-05','2021-06-05',
                     '2021-11-01','2021-11-02','2021-11-11','2021-12-12']),
  'lower_window': 0,
  'upper_window': 1,
})
```

```python
# define the model
model = Prophet(changepoint_range=0.9,n_changepoints=60, changepoint_prior_scale=0.20,
                yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=True, 
                seasonality_prior_scale=10, seasonality_mode='multiplicative',
                holidays=holidays, holidays_prior_scale=10,
                mcmc_samples=0, interval_width=0.8, uncertainty_samples=1000) #初始化;yearly_seasonality周期为年的季节性
```

```python
# fit the model
model.fit(train)
```

```python
def datelist_gen(start,end):
    datelist = list()
    #start = '2020-01-01'
    #end = '2021-12-31'
    datestart = datetime.datetime.strptime(start, '%Y-%m-%d')
    dateend = datetime.datetime.strptime(end, '%Y-%m-%d')
    for x in range((dateend - datestart).days + 1):
        new = datestart + datetime.timedelta(days=x)
        date = str(new)[:-9]
        datelist.append([date])
    datelist = DataFrame(datelist)
    datelist.columns = ['ds']
    datelist['ds'] = to_datetime(datelist['ds'])
    return datelist
```

```python
#future = datelist_gen('2021-04-01','2021-12-31')
#future
future = model.make_future_dataframe(periods=275)
#future
```

```python
forecast = model.predict(future)

print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
```

```python
# plot forecast
fig = model.plot(forecast)
ax = fig.add_subplot(111)
ax.plot(df_log['ds'][(df['ds']>='2021-04-01')&(df['ds']<='2021-06-30')].astype('datetime64'), 
        df_log['y'][(df['ds']>='2021-04-01')&(df['ds']<='2021-06-30')], 'g.')
plt.savefig('../gen_data/prophet_fit_with_holiday.jpg')
plt.show()
```

```python
#数据还原
#df['y'] = np.exp(df['y'])-1
predict_df = forecast[['ds','yhat']]
predict_df['yhat'] = np.exp(predict_df['yhat'])-1
```

```python
#计算MAPE值
MAPE_score = mean_absolute_percentage_error(df['y'][-91:], predict_df['yhat'][(predict_df['ds']>='2021-04-01')&(predict_df['ds']<='2021-06-30')])
print('Forecast MAPE Score: %.2f MAPE' % MAPE_score)
```

```python
merge(df[(df['ds']>='2021-06-01')&(df['ds']<='2021-06-30')], 
      predict_df[(predict_df['ds']>='2021-06-01')&(predict_df['ds']<='2021-06-30')], on='ds')
```

```python
y_true_date = datelist_gen('2021-04-01','2021-06-30')
y_predict_date = datelist_gen('2021-04-01','2021-12-30')
y_true = df['y'][(df['ds']>='2021-04-01')&(df['ds']<='2021-06-30')]
y_predict = predict_df['yhat'][(predict_df['ds']>='2021-04-01')&(predict_df['ds']<='2021-12-30')]
```

```python
plt.figure(figsize=(24,12))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel("sales_num", fontsize=22)
plt.grid(True, which='both', linestyle='-')
plt.plot(y_true_date, y_true, label='y_true')
plt.plot(y_predict_date, y_predict, label='y_predict')
plt.xticks(rotation=15)
plt.legend(fontsize=20)
plt.savefig('../gen_data/prophet_result_with_holiday.jpg')
plt.show()
```

### 年月和日分析

```python
date = data['take_at'].str.split(' ').str[0]
data['year-month-day']=date
data['year-month'] = date.str.split('-').str[0]+'-'+date.str.split('-').str[1]
```

```python
data_last = data[['status','payment','year-month-day','year-month']]
data_last=data_last[data_last['status']==6]
data_last.head()
```

```python
# 看日期情况
date_= data_last.groupby('year-month-day')['year-month-day'].count()
date_
```

```python
data_day = data_last.groupby('year-month-day')
list(data_day)[0][1]
```

```python
data_day = data_last.groupby('year-month-day')
# 每一天的GMV数据(status=6)
gmv_day= []
for i in tqdm(range(len(list(data_day)))):
    gmv_1=list(data_day)[i][1]['payment'].sum()
    gmv_day.append(gmv_1)
gmv_day
```

```python
# 由于一共547天,不是7的整数倍，所以取到前546天的数据作为周的数据来源
gmv_week=[]
m,n=0,0
for i in range(len(gmv_day)):
    m+=gmv_day[i]
    n+=1
    if n%7==0:
        gmv_week.append(m)
        n=0
        m=0
print(len(gmv_week))
gmv_week
```

```python
data_month = data_last.groupby('year-month')
# 每一月的GMV数据(status=6)
gmv_month= []
for i in tqdm(range(len(list(data_month)))):
    gmv_2=list(data_month)[i][1]['payment'].sum()
    gmv_month.append(gmv_2)
gmv_month
```

```python
print(f"gmv_day_std:{np.std(gmv_day,ddof=1):.4f}----")
print(f"gmv_week_std:{np.std(gmv_week,ddof=1):.4f}----")
print(f"gmv_month_std:{np.std(gmv_month,ddof=1):.4f}----")
```

##创建次日比本日高的数据（创建次周比本周高的数据，创建次月比本月高的数据）

```python
# 创建次日比本日高的数据
up_day=[]
for i in range(len(gmv_day)):
    if i>=len(gmv_day)-1:
        continue
    else:
        up_day.append(gmv_day[i+1]-gmv_day[i])

# 创建次周比本周高的数据
up_week=[]
for i in range(len(gmv_week)):
    if i>=len(gmv_week)-1:
        continue
    else:
        up_week.append(gmv_week[i+1]-gmv_week[i])
        
# 创建次月比本月高的数据
up_month=[]
for i in range(len(gmv_month)):
    if i>=len(gmv_month)-1:
        continue
    else:
        up_month.append(gmv_month[i+1]-gmv_month[i])     
print(f"-----up_day数据长度：{len(up_day)}-----up_week数据长度：{len(up_week)}----up_month数据长度：{len(up_month)}")
```

```python
# 画图
from pyecharts.charts import Line
from pyecharts import options as opts
# 导入输出图片工具# 使用snapshot-selenium 渲染图片
from pyecharts.render import make_snapshot
from snapshot_selenium import snapshot


# month_list=pd.period_range('2020-01', periods=18, freq='M')
# month_list = [str(month_list[i]) for i in range(len(month_list))]

def line_charts(data,x_name,y_name,title) -> Line:
    line = (
        Line()
        .add_xaxis(month_list)
        .add_yaxis(
            "GMV值",
            data
        )
        .set_global_opts(title_opts=opts.TitleOpts(title=title)
                         ,yaxis_opts=opts.AxisOpts(name=y_name)
                         ,xaxis_opts=opts.AxisOpts(name=x_name),)
         .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False), # 是否显示标签
        )
    )
    return line
# # 天的gmv
line_day = line_charts(gmv_day,x_name='day',y_name='gmv_num',title='gmv_day_analyse')
line_day.render("gmv_day1分析.html")

# # # 周的GMV
line_week = line_charts(gmv_week,x_name='week',y_name='gmv_num',title='gmv_week_analyse')
line_week.render("gmv_week1分析.html")
# # 输出保存为图片
# make_snapshot(snapshot, "gmv_week分析.html", "gmv_week分析.png")

# # 月的GMV
line_month = line_charts(gmv_month,x_name='month',y_name='gmv_num',title='gmv_month_analyse')
line_month.render("gmv_month1分析.html")
# # 输出保存为图片
# make_snapshot(snapshot, "gmv_month分析.html", "gmv_month分析.png")
```

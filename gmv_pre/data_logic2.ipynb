```python
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error 

def mape(y_true, y_pred):
    """
    参数:
    y_true -- 测试集目标真实值
    y_pred -- 测试集目标预测值
    返回:
    mape -- MAPE 评价指标
    """
    n = len(y_true)
    mape = sum(np.abs((y_true - y_pred)/y_true))/n*100
    return mape
```

```python
# 定义训练函数
x_train = train.values
print(x_train)
x_test = test.values
print(x_test)
def rmsle_cv(model):
    model.fit(x_train, y_train)
    y_hat1 = model.predict(x_train)
    y_hat2 = model.predict(x_test)
    mse = np.average((y_hat2 - np.array(y_test)) ** 2)  # Mean Squared Error
    rmse = np.sqrt(mse)  # Root Mean Squared Error
    mae = mape(y_test,y_hat2)
    return mae , rmse,y_hat1,y_hat2
```

### 模型 定义

```python
# RobustScaler() 后lasso
lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.000005, random_state=1))
# RobustScaler() 后enet
ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.000005, l1_ratio=0.95, random_state=1))
```

```python
mae_lasso,rmse_lasso,y_lasso_hat_train,y_lasso_hat_test = rmsle_cv(lasso)
print(f"Lasso_Rmse: {rmse_lasso:.4f}----Lasso_MAPE:{mae_lasso:.6f}%")

mae_Enet,rmse_Enet, y_enet_hat_train,y_enet_hat_test = rmsle_cv(ENet)
print(f"Enet_Rmse: {rmse_Enet:.4f}----Enet_MAPE:{mae_Enet:.6f}%")
```

```python
# 平均模型
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, models):
        self.models = models
        
    # we define clones of the original models to fit the data in
    def fit(self, X, y):
        self.models_ = [clone(x) for x in self.models]
        
        # Train cloned base models
        for model in self.models_:
            model.fit(X, y)

        return self
    
    # Now we do the predictions for cloned models and average them
    def predict(self, X):
        predictions = np.column_stack([
            model.predict(X) for model in self.models_
        ])
        return np.mean(predictions, axis=1)
# 开始训练
averaged_models = AveragingModels(models = (ENet,lasso))                       
avg_mape,avg_rmse,y_avg_hat_train,y_avg_hat_test = rmsle_cv(averaged_models)
print(f"Enet_Rmse: {avg_rmse:.4f}----Enet_MAPE:{avg_mape:.6f}%")    
```

```python
# 堆叠模型
class StackingAveragedModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, base_models, meta_model, n_folds=5):
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    # We again fit the data on clones of the original models
    def fit(self, X, y):
        self.base_models_ = [list() for x in self.base_models]
        self.meta_model_ = clone(self.meta_model)
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        
        # Train cloned base models then create out-of-fold predictions
        # that are needed to train the cloned meta-model
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models)))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                instance = clone(model)
                self.base_models_[i].append(instance)
                instance.fit(X[train_index], y[train_index])
                y_pred = instance.predict(X[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
                
        # Now train the cloned  meta-model using the out-of-fold predictions as new feature
        self.meta_model_.fit(out_of_fold_predictions, y)
        return self
   
    # Do the predictions of all base models on the test data and use the averaged predictions as 
    # meta-features for the final prediction which is done by the meta-model
    def predict(self, X):
        meta_features = np.column_stack([
            np.column_stack([model.predict(X) for model in base_models]).mean(axis=1)
            for base_models in self.base_models_ ])
        return self.meta_model_.predict(meta_features)
# 开始训练
stacked_averaged_models = StackingAveragedModels(base_models = (lasso,lasso),
                                                 meta_model = lasso)
score_mape,score_rmse,y_stacking_hat_train,y_stacking_hat_test = rmsle_cv(stacked_averaged_models)
print(f"stack_Rmse: {score_rmse:.4f}----stack_MAPE:{score_mape:.4f}%")
```
